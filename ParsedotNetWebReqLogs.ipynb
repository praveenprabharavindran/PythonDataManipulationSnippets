```python
import re
from google.colab import files

uploaded = files.upload()
file_name = list(uploaded.keys())[0]

multiline_string = uploaded[file_name].decode('utf-8')
```

```python
pattern = r"(?P<method>GET|POST|DELETE|PUT)\s+(?P<url>https?:\S+)\s+.*?after\s+(?P<receiveTime>\d+\.\d+)ms.*\s+.*after\s+(?P<endTime>\d+\.\d+)ms\s-\s(?P<status>\d+)"

matches = re.finditer(pattern, multiline_string, re.MULTILINE)

```

```python
table_header = "|".join(["Url", "Method", "ReceiveTime", "EndTime", "Status"])
print(table_header)
print("-" * len(table_header))

for match in matches:
    url = match.group("url")
    method = match.group("method")
    receive_time = match.group("receiveTime")
    end_time = match.group("endTime")
    status = match.group("status")

    table_row = "|".join([url, method, receive_time, end_time, status])
    print(table_row)
```
